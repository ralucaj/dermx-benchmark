```python
import glob
import pandas as pd
from pathlib import Path
from sklearn.metrics import classification_report, f1_score, recall_score
from IPython.display import display, HTML

import warnings
warnings.filterwarnings("ignore")

base_paths = [
    "/home/ubuntu/hot-store/final/resnet",
    "/home/ubuntu/hot-store/final/efficientnet",
    "/home/ubuntu/hot-store/final/inceptionresnet",
    "/home/ubuntu/hot-store/final/inception",
    "/home/ubuntu/hot-store/final/mobilenetv1",
    "/home/ubuntu/hot-store/final/mobilenetv2",
    "/home/ubuntu/hot-store/final/nasnetmobile",
    "/home/ubuntu/hot-store/final/resnetv2",
    "/home/ubuntu/hot-store/final/vgg",
    "/home/ubuntu/hot-store/final/xception",
]

for base_path in base_paths:
    model_name = base_path.split("/")[-1]
    model_preds = glob.glob(base_path + f"/{model_name}*_preds.csv")
    model_comparison_dict = {}

    for model_pred in model_preds:
        model_preds_df = pd.read_csv(Path(model_pred))
        model_comparison_dict[Path(model_pred).stem] = {}
        for label, diagnosis in enumerate(['acne', 'actinic_keratosis', 'psoriasis_no_pustular', 'seborrheic_dermatitis', 'vitiligo', 'wart']):
            model_comparison_dict[Path(model_pred).stem][diagnosis] = {}
            y_actual = model_preds_df['actual'] == label
            y_pred = model_preds_df['pred'] == label
            model_comparison_dict[Path(model_pred).stem][diagnosis]['F1 score'] = f1_score(
                y_actual,
                y_pred,            )
            model_comparison_dict[Path(model_pred).stem][diagnosis]['sensitivity'] = recall_score(
                y_actual,
                y_pred,
            )
            model_comparison_dict[Path(model_pred).stem][diagnosis]['specificity'] = recall_score(
                y_actual,
                y_pred,
                pos_label=0
            )
        
            model_comparison_df = pd.DataFrame.from_dict(model_comparison_dict, orient='index')    
            model_comparison_df = pd.DataFrame.from_dict(model_comparison_dict).stack().to_frame()
            model_comparison_df = pd.DataFrame(model_comparison_df[0].values.tolist(), index=model_comparison_df.index).reset_index()
            model_comparison_df = model_comparison_df.groupby('level_1').mean()
    display(HTML(model_comparison_df.head().to_html()))
    display(HTML(model_comparison_df.describe().to_html()))

#     print(f'{model_name}: {model_comparison_df["f1-score"].mean()} \pm {model_comparison_df["f1-score"].std()}')
```
