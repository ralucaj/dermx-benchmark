## Masks

```python
import os
import pandas as pd
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

base_path = '/home/ubuntu/hot-store/dermx/masks/masks/'
target_path = '/home/ubuntu/hot-store/dermx_masks/'
```

```python
annotations = pd.read_csv('/home/ubuntu/hot-store/dermx/masks/instance_mask_annotations.csv').set_index('mask_id')
annotations
```

```python
mask_files = os.listdir('/home/ubuntu/hot-store/dermx/masks/masks')
mask_files
```

```python
# Extract masks info
masks_dict = {
    'mask_file': [],
    'derm': [],
    'image': [],
    'characteristic': []
}
for mask_file in mask_files:
    derm, image, _ = mask_file[:-4].split('_')
    characteristic = annotations.loc[mask_file[:-4]]['class_name']
    masks_dict['mask_file'].append(mask_file)
    masks_dict['derm'].append(derm)
    masks_dict['image'].append(image)
    masks_dict['characteristic'].append(characteristic)
masks_df = pd.DataFrame.from_dict(masks_dict)
masks_df
```

```python
def fuzzy_union_images(base_path, filenames, target_filename):
    images = np.array([np.asarray(Image.open(base_path + filename)) for filename in filenames])
    images = images.sum(axis=0)
    images = images / images.max() * 256
    images = Image.fromarray(images).convert("L")
    images.save(target_filename)
```

```python
# Build per derm union
for derm in masks_df['derm'].unique():
    for image in masks_df['image'].unique():
        mask_files = masks_df[(masks_df['derm'] == derm) & (masks_df['image'] == image)].mask_file.values
        if len(mask_files):
            fuzzy_union_images(base_path, mask_files, target_path + f'per_derm/{derm}_{image}.png')
```

```python
# Build full union
for image in masks_df['image'].unique():
    mask_files = masks_df[masks_df['image'] == image].mask_file.values
    if len(mask_files):
        fuzzy_union_images(base_path, mask_files, target_path + f'per_image/{image}.png')
```

```python
# Build per char union
for characteristic in masks_df['characteristic'].unique():
    for image in masks_df['image'].unique():
        mask_files = masks_df[(masks_df['characteristic'] == characteristic) & (masks_df['image'] == image)].mask_file.values
        if len(mask_files):
            fuzzy_union_images(base_path, mask_files, target_path + f'per_characteristic/{characteristic}_{image}.png')
```
