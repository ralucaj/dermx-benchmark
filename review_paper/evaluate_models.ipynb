```python
import glob
import pandas as pd
from pathlib import Path
from sklearn.metrics import classification_report
from IPython.display import display, HTML

import warnings
warnings.filterwarnings("ignore")

base_paths = [
    "/home/ubuntu/hot-store/final/resnet",
    "/home/ubuntu/hot-store/final/efficientnet",
    "/home/ubuntu/hot-store/final/inceptionresnet",
    "/home/ubuntu/hot-store/final/inception",
    "/home/ubuntu/hot-store/final/mobilenetv1",
    "/home/ubuntu/hot-store/final/mobilenetv2",
    "/home/ubuntu/hot-store/final/nasnetmobile",
    "/home/ubuntu/hot-store/final/resnetv2",
    "/home/ubuntu/hot-store/final/vgg",
    "/home/ubuntu/hot-store/final/xception",
]

for base_path in base_paths:
    model_preds = glob.glob(base_path + "/dermx*_preds.csv")
    model_comparison_dict = {}

    for model_pred in model_preds:
        model_preds_df = pd.read_csv(Path(model_pred))
        model_comparison_dict[Path(model_pred).stem] = classification_report(
            model_preds_df['actual'],
            model_preds_df['pred'],
            labels=[0, 1, 2, 3, 4, 5],
            target_names=['acne', 'actinic_keratosis', 'psoriasis_no_pustular', 'seborrheic_dermatitis', 'vitiligo', 'wart'],
            output_dict=True
        )['macro avg']

    model_comparison_df = pd.DataFrame.from_dict(model_comparison_dict, orient='index')
    display(HTML(model_comparison_df.head().to_html()))
    
    print(f'{base_path.split("/")[-1]} precision: {model_comparison_df["precision"].mean()} \pm {model_comparison_df["precision"].std()}')
    print(f'{base_path.split("/")[-1]} recall: {model_comparison_df["recall"].mean()} \pm {model_comparison_df["recall"].std()}')
    print(f'{base_path.split("/")[-1]} f1: {model_comparison_df["f1-score"].mean()} \pm {model_comparison_df["f1-score"].std()}')
    
```
